```python
#What is the total price for each category of maker?

#groupby

Auto.groupby("make")["price"].sum()
```

```python
#Fuel type wise total price
```

```python
Auto.groupby("fuel-type")["price"].sum()
```

```python
#Average price of car by body-style wise

print("Average stype as per body-style: ",Auto.groupby("body-style")["price"].mean())
```

```python
#Min price of car by engine type

print("Min price of car by engine type:",Auto.groupby("engine-type")["price"].min())
```

```python
#to check average highway-mpg of all fuel type

Auto.groupby("fuel-type")["highway-mpg"].mean()
```

```python
#find total price of cars body style and fuel type wise

Auto.groupby(["body-style","fuel-type"])["highway-mpg"].sum()
```

```python
#What is average highway-mpg given by each maker & fuel type under them?

Auto.groupby(["make","fuel-type"])["highway-mpg"].mean()
```
