```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
Titanic=sns.load_dataset("titanic")
```

```python
Titanic
```

```python
sns.displot(Titanic["age"])
plt.show()
```

```python
#stacked

#Chart showing age wise distribution and class wise division in each group

sns.displot(Titanic,x="age",hue="class")
plt.show()
```

```python
#Chart showing age wise male female
sns.displot(Titanic,x="age",hue="sex")
plt.show()
```

```python

sns.displot(Titanic,x="age",hue="survived",palette="flare")
plt.show()
```

```python
sns.pairplot(Titanic)
plt.show()
```

```python
sns.catplot(data=Titanic,y="survived",x="sex",hue="class",kind="violin")
plt.show()
```

```python
#create countplot for survived and non survived people
sns.countplot(data=Titanic,x="survived",palette="mako")
plt.show()
```

```python
#create a countplot with barchart of different class wise count
sns.countplot(data=Titanic,y="class",palette="mako")
plt.show()
```

```python
#create pie chart to check how many percent were male and female
color = sns.color_palette("Set2")
Titanic["sex"].value_counts().plot(kind="pie",autopct="%.0f%%",colors=color)
plt.show()
```
